```python
import json
import numpy as np
import os
from pathlib import Path

manifest = json.load(open(Path(os.getcwd()).parent / "geom_drugs_conformers/manifest.json", "r"))

boltzmann_lengths = []
boltzmann_sums= []
for record in manifest:
    boltzmann_lengths.append(len(record['boltzmann_weights']))
    boltzmann_sums.append(sum(record['boltzmann_weights']))

np.mean(boltzmann_lengths)
```

```python
np.min(boltzmann_lengths), np.max(boltzmann_lengths)
```

```python
np.min(boltzmann_sums), np.max(boltzmann_sums)
```
